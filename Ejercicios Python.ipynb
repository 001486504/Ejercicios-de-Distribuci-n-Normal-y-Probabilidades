```python
#Ejercicio 1
import numpy as np
import matplotlib.pyplot as plt

media = 60
desviacion_estandar = 15
muestra = np.random.normal(loc=media, scale=desviacion_estandar, size=1000)

# Crear el histograma de la muestra
plt.figure(figsize=(10, 6))
plt.hist(muestra, bins=30, color='skyblue', edgecolor='black')
plt.title("Histograma de la Muestra con Distribución Normal")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()

# Calcular
media_muestra = np.mean(muestra)
desviacion_estandar_muestra = np.std(muestra)
print("Media:", media_muestra)
print("Desviación Estándar:", desviacion_estandar_muestra)

#---------------------------------------------------------------------------------------------#
#  Resumen de su Funcionamiento                                                               #
#                                                                                             #
# - Genera una muestra normal de 1000 valores con media y desviación estándar especificadas.  #
# - Dibuja un histograma de la muestra para visualizar la distribución.                       #
# - Calcula y muestra la media y desviación estándar de la muestra.                           #
#                                                                                             #
#---------------------------------------------------------------------------------------------#
```

```python
#Ejercicio 3
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generar una muestra de 500 observaciones con media = 75 y desviación estándar = 8
media = 75
desviacion_estandar = 8
muestra = np.random.normal(loc=media, scale=desviacion_estandar, size=500)

# Calcular la media de la muestra y el intervalo de confianza al 95%
media_muestra = np.mean(muestra)
error_estandar = stats.sem(muestra)  # Error estándar de la media
intervalo_confianza = stats.t.interval(0.95, df=len(muestra)-1, loc=media_muestra, scale=error_estandar)

# Crear la gráfica de la muestra con el intervalo de confianza
plt.figure(figsize=(12, 6))
plt.hist(muestra, bins=30, color='lightgreen', edgecolor='black', alpha=0.7, label="Muestra")
plt.axvline(intervalo_confianza[0], color='red', linestyle='--', label="Límite Inferior IC 95%")
plt.axvline(intervalo_confianza[1], color='red', linestyle='--', label="Límite Superior IC 95%")
plt.axvline(media_muestra, color='blue', linestyle='-', label="Media Muestral")
plt.title("Histograma de la Muestra con Intervalo de Confianza al 95% para la Media")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()

# Imprimir los límites del intervalo de confianza
print("Intervalo de confianza al 95% para la media:", intervalo_confianza)

#---------------------------------------------------------------------------------------------#
#  Resumen de su Funcionamiento                                                               #
#                                                                                             #
# - Genera una muestra de 500 valores con media y desviación estándar dadas.                  #
# - Calcula la media y el intervalo de confianza al 95% para la media.                        #
# - Grafica el histograma de la muestra e indica el intervalo de confianza y la media.        #
# - Imprime el intervalo de confianza al 95% para la media de la muestra.                     #
#                                                                                             #
#---------------------------------------------------------------------------------------------#
```

```python
#ejercicio 5
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de 100 valores con media = 100 y desviación estándar = 20
media = 100
desviacion_estandar = 20
muestra = np.random.normal(loc=media, scale=desviacion_estandar, size=100)

# Calcular el percentil 90 de la muestra
percentil_90 = np.percentile(muestra, 90)
print("El percentil 90 de la muestra es:", percentil_90)

# Crear el histograma de la muestra
plt.figure(figsize=(10, 6))
plt.hist(muestra, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Añadir línea del percentil 90
plt.axvline(percentil_90, color='red', linestyle='dashed', linewidth=2, label=f'Percentil 90: {percentil_90:.2f}')

# Añadir título y etiquetas
plt.title("Histograma de la Muestra con Percentil 90")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.legend()

# Mostrar el gráfico
plt.show()
#--------------------------------------------------------------------------------------------------------------#
#  Resumen de su Funcionamiento                                                                                #
#                                                                                                              #
# - Generación de muestra: Crea una muestra de 100 valores con parámetros definidos.                           #
# - Cálculo del percentil 90: Determina el valor que corresponde al percentil 90 de la muestra.                #
# - Visualización: Dibuja un histograma de la muestra y marca el percentil 90 con una línea roja discontinua.  #
#                                                                                                              #
#--------------------------------------------------------------------------------------------------------------#
```

```python
#ejercicio 7
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de 1000 valores de una distribución normal estándar
muestra = np.random.normal(loc=0, scale=1, size=1000)

# Calcular el porcentaje de valores entre -1 y 1
valores_entre_1_y_menos_1 = np.sum((muestra >= -1) & (muestra <= 1))
porcentaje = (valores_entre_1_y_menos_1 / len(muestra)) * 100

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(muestra, bins=30, color='skyblue', edgecolor='black', density=True, alpha=0.6)

# Generar valores para la curva de densidad
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, color='blue', label='Distribución normal estándar')

# Sombrear el área entre -1 y 1
x_fill = np.linspace(-1, 1, 100)
y_fill = stats.norm.pdf(x_fill, 0, 1)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.4, label='Valores entre -1 y 1')

# Mostrar el porcentaje en el gráfico
plt.text(1.5, 0.25, f"Porcentaje entre -1 y 1: {porcentaje:.2f}%", fontsize=12, color="black")

# Personalizar el gráfico
plt.title("Distribución normal estándar y porcentaje de valores entre -1 y 1")
plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
#-----------------------------------------------------------------------------------------------------------------------#
#  Resumen de su Funcionamiento                                                                                         #
#                                                                                                                       #
# - Generar muestra: Crea una muestra de 1000 valores de una distribución normal estándar.                              #
# - Calcular porcentaje: Calcula el porcentaje de valores que caen entre -1 y 1 en la muestra.                          #
# - Dibujar histograma: Muestra el histograma de la muestra y la curva de densidad de la distribución normal estándar.  #
# - Resaltar área entre -1 y 1: Sombrea el área entre -1 y 1 en la curva para visualizar el porcentaje.                 #
# - Mostrar el porcentaje: Añade el porcentaje calculado en el gráfico para su referencia.                              #
#-----------------------------------------------------------------------------------------------------------------------#
```

```python
#ejercicio 9
import numpy as np
import matplotlib.pyplot as plt

# Generar una muestra de 150 valores con media = 60 y desviación estándar = 10
muestra = np.random.normal(loc=60, scale=10, size=150)

# Transformar los valores de la muestra para que tengan media 0 y desviación estándar 1
muestra_transformada = (muestra - np.mean(muestra)) / np.std(muestra)

# Verificar los estadísticos de la muestra transformada
media_transformada = np.mean(muestra_transformada)
desviacion_transformada = np.std(muestra_transformada)

# Crear el histograma de la muestra transformada
plt.hist(muestra_transformada, bins=25, color='skyblue', edgecolor='black', alpha=0.7)

# Añadir el título y las etiquetas
plt.title('Distribución de la Muestra Transformada', fontsize=12)
plt.xlabel('Valor', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

# Generar valores para la curva de densidad
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, color='blue', label='Distribución normal estándar')

# Sombrear el área entre -1 y 1
x_fill = np.linspace(-1, 1, 100)
y_fill = stats.norm.pdf(x_fill, 0, 1)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.4, label='Valores entre -0.00 y 1.00')


# Añadir las estadísticas de la muestra transformada en el gráfico
plt.text(0.05, 15, f"Media: {media_transformada:.2f}", fontsize=12, color='red')
plt.text(0.05, 14, f"Desviación estándar: {desviacion_transformada:.2f}", fontsize=12, color='red')

# Mostrar el gráfico
plt.show()

print(f"Media de la muestra transformada: {media_transformada:.2f}")
print(f"Desviación estándar de la muestra transformada: {desviacion_transformada:.2f}")
#-----------------------------------------------------------------------------------------------------------------------#
#  Resumen de su Funcionamiento                                                                                         #
#                                                                                                                       #
# - Genera una muestra con una distribución normal de 150 valores con media 60 y desviación estándar 10.                #
# - Transforma la muestra para que tenga una media de 0 y una desviación estándar de 1.                                 #
# - Crea un histograma de la muestra transformada.                                                                      #
# - Muestra la media y la desviación estándar de la muestra transformada dentro del gráfico.                            #
#-----------------------------------------------------------------------------------------------------------------------#
```
